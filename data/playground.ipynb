```python
from pyspark.sql import SparkSession


spark = SparkSession.builder \
    .appName("MyApp") \
    .config("spark.jars", "/home/jovyan/data/mysql-connector-j-8.3.0.jar") \
    .config("spark.jars", "/home/jovyan/data/postgresql-42.7.1.jar") \
    .getOrCreate()
```

```python
df_orders = spark.read \
    .format("jdbc") \
    .option("url", "jdbc:mysql://mysql:3306/masterclass") \
    .option("user", "mysqluser") \
    .option("password", "mysqlpw") \
    .option("dbtable", "orders") \
    .option("inferSchema", "true") \
    .load()

df_orders.show()
```
